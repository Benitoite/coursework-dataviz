```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import warnings
import calendar as cal
import matplotlib.font_manager as fm
import matplotlib
warnings.filterwarnings("ignore")
```

```python
%matplotlib inline

plt.rcdefaults()
font_path = 'C:\\Windows\\Fonts\\Humor-Sans.ttf'
font_prop = fm.FontProperties(fname=font_path, size=14)

def weekdays(weekday):
    start = [d for d in cal.day_name].index(weekday)
    return [cal.day_name[(i+start) % len(cal.day_name)] for i in range(len(cal.day_name))]

days = weekdays(cal.day_name[0])

# Make a dataframe with days before weekend values
until = pd.DataFrame()
for i in days:
    until.at[days[days.index(i)],0]=(5-(days.index(i)))

# Initialize a plot in the XKCD style
with plt.xkcd():
    title   ='Days Until Weekend'
    
    fig, ax = plt.subplots(figsize=[11,7])
    ind     = days
    x       = until.index
    h       = until.values
    my_cmap = plt.get_cmap("RdYlGn_r")
    color   = np.linspace(256,0,num=len(cal.day_name)).astype(int)
    
    # Set the minimum value to zero
    until[until < 0] = 0
    
    # Make a bar chart
    ax.bar(ind,until[0],width=.6,color=my_cmap(color))
    plt.xticks(rotation = 22)
    ax.set_xticklabels(labels=days, fontproperties=font_prop)
    ax.set_yticklabels(labels=range(6), fontproperties=font_prop)
    
    plt.title('Days Until the Weekend', fontproperties=font_prop)

    # Add a colorbar
    colorbar = fig.colorbar(cm.ScalarMappable(cmap=my_cmap), ax=ax)
    colorbar.ax.set_yticklabels(['undefined', 'almost', 'gettin\' over it', 'ho hum', 'so far away', cal.day_name[0]], fontproperties=font_prop) 

plt.show()
plt.close()
```
