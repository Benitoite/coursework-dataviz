```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
days  = ['monday','tuesday','wednesday','thursday','friday','saturday','sunday']
until = pd.DataFrame()

# Make a dataframe with days before weekend values
for i in days:
    until.at[days[days.index(i)],0]=(5-(days.index(i)))

# Initialize a plot in the XKCD style
with plt.xkcd():
    title   ='Days Until Weekend'
    fig, ax = plt.subplots(figsize=[11,7]) 
    ind     = days
    x       = until.index
    h       = until.values
    my_cmap = plt.get_cmap("inferno")
    color   = np.linspace(256,0,num=7).astype(int)
    
    # Set the minimum value to zero
    until[until < 0] = 0
    
    # Make a bar chart
    ax.bar(ind,until[0],width=.6,color=my_cmap(color))
    plt.xticks(rotation = 22)
    plt.title('Days Until the Weekend')
    
    # Add a colorbar
    colorbar = fig.colorbar(cm.ScalarMappable(cmap=my_cmap), ax=ax)
    colorbar.ax.set_yticklabels(['undefined', 'almost', 'gettin\' over it', 'ho hum', 'so far away', 'monday']) 
    
plt.show()
plt.close()
```
